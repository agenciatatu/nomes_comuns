```python
import json
import requests
import time
from random import randint

#Lista todas as UFs do Brasil
listaUF = ["AC","AL","AP","AM","BA","CE","DF","ES","GO","MA","MT","MS","MG","PA","PB","PR","PE","PI","RJ","RN","RS","RO","RR","SC","SP","SE","TO"]

#Lista todos os anos disponiveis
listaAno = [2015,2016,2017,2018,2019]

listacompleta = []


#Itera em cada estado
for uf in listaUF:
    time.sleep(randint(0,15))
    
    #Itera em cada ano disponível
    for ano in listaAno:
        print(ano)
        print(uf)
        
        #Faz uma solicitação à API do registrocivil.org.br com o ano e estado pretendido
        response = requests.get("https://transparencia.registrocivil.org.br/api/record/all-name?start_date="+str(ano)+"-01-01&end_date="+str(ano)+"-12-31&translate=1&state="+uf)
        
        time.sleep(0.1)
        
        print(response)
        print(response.text)
        
        #print(response.text)
        resposta = json.loads(response.text)
        resposta = resposta["data"]
        
        #Adiciona ano e estado a cada item do dicionário
        for item in resposta:
            item["ano"] = ano
            item["estado"] = uf
            
        #Pega só o primeiro item, que é o nome mais registrado, por ano e estado
        try:
            print(resposta[0])
            listacompleta.append(resposta[0].copy())
        except:
            pass
        #Adiciona cada item à lista "listacompleta"
        #for item in resposta:
            #print(item)
            #listacompleta.append(item.copy())
     
```

```python
import pandas as pd

#Transforma a lista completa em um dataframe e salva em csv
df = pd.DataFrame(listacompleta)
```

```python
df.to_csv("Registro_civil_mais_do_ano.csv")
```
